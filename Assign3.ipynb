**PART 1**

Numerical values are values that can be divided into categories or groups despite being represented as integer numbers. They take limited and fixed number of possible values. 

As an example, that could be a rating of a movie. 
Rating = (5:Excellent, 4:Very good, 3:Good, 2:Average, 1:Poor)

```python
import pandas as pd 
```

```python
data = {
    'movie_name' : ["Scent of a woman", "Prestige", "Insomnia", "Sully", "XXXX"], 
    'rating' : ['Excellent', 'Very Good', 'Good', 'Average', 'Poor'] 
}
df = pd.DataFrame(data)
```

```python
rating_mapping = {
    'Poor': 1,
    'Average': 2,
    'Good': 3,
    'Very Good': 4,
    'Excellent': 5
}

df['rating'] = df['rating'].map(rating_mapping)

df
```
